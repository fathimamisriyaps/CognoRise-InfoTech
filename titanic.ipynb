```python
df.drop(['PassengerId','Cabin','Ticket','Name'],axis=1,inplace=True)
```

handling missing values

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
df['Fare']=df['Fare'].fillna(df['Fare'].mean())
```

```python
#encoding categorical variables
df['Embarked'] = df['Embarked'].map({'Q': 0, 'S': 1, 'C': 2}).astype(int)
df['Sex'] = df['Sex'].map({'female': 1, 'male': 0}).astype(int)
```

```python
df
```

```python
y=df['Survived'].values
y
```

```python
x=df.iloc[:,1:].values
x
```

```python
x.ndim
```

splitting the dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

normalisation

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

Model building

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(criterion='entropy')
k_model=KNeighborsClassifier(n_neighbors=9,weights='uniform')
n_model=GaussianNB()
s_model=SVC()
lst_model=[k_model,n_model,s_model]
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,ConfusionMatrixDisplay
for i in lst_model:
   print('model name',i)
   i.fit(x_train,y_train)
   y_pred=i.predict(x_test)
   score=accuracy_score(y_test,y_pred)
   report=classification_report(y_test,y_pred)
   mat=confusion_matrix(y_test,y_pred)
   dis=ConfusionMatrixDisplay(mat)    #object created to cnfusion matrix displaymat

   print(y_pred)
   print('*'*100)
   print(score)
   print('*'*100)
   print(mat)
   print('*'*100)
   print(report)

   dis.plot()
```
